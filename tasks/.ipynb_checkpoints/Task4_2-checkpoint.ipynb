Явка (%) по всем регионам, результат отсортирован по убыванию.

```python
presence = dict()
# количество заполненных бюллютеней
m1_col = 'Число недействительных избирательных бюллетеней'
m2_col = 'Число действительных избирательных бюллетеней'
# общее количество бюллютеней
n_col = 'Число избирателей, включенных в список избирателей ' 
```

```python
for region in df['Регион'].unique():
    presence[region] = df[df['Регион'] == region][[m1_col, m2_col]].sum().sum() / df[df['Регион'] == region][n_col].sum()
presence_df = pd.Series(presence)
```

```python
presence_df.sort_values(ascending=False, inplace=True)
presence_df[:10]
```

```python
candidates = ['Бабурин Сергей Николаевич', 'Грудинин Павел Николаевич', 'Жириновский Владимир Вольфович', 
'Путин Владимир Владимирович', 'Собчак Ксения Анатольевна', 'Сурайкин Максим Александрович', 
'Титов Борис Юрьевич', 'Явлинский Григорий Алексеевич'] 
candidate = 'Жириновский Владимир Вольфович' 

res = df[df[candidate] > 300][candidate]/df[df[candidate] > 300].sum(axis=1) 
max_index = res.idxmax() 
print(f'{candidate} набрал {df.iloc[max_index][candidate]} голосов в регионе {df.iloc[max_index]["Регион"]}, ТИК {df.iloc[max_index]["ТИК"]}, УИК №{df.iloc[max_index]["Номер УИК"]}')
```

Регион, где разница между ТИК с наибольшей явкой и наименьшей максимальна

```python
presence = dict()
# количество заполненных бюллютеней
m1_col = 'Число недействительных избирательных бюллетеней'
m2_col = 'Число действительных избирательных бюллетеней'
# общее количество бюллютеней
n_col = 'Число избирателей, включенных в список избирателей '

max_delta = {'region': None, 'delta': -1}
for region in df['Регион'].unique():
    presence_region = df[df['Регион'] == region][[m1_col, m2_col]].sum().sum() / df[df['Регион'] == region][n_col]
    delta = presence_region.max() - presence_region.min()
    if delta > max_delta['delta']:
        max_delta['delta'] = delta
        max_delta['region'] = region
max_delta
```

Дисперсия по явке для каждого региона (учитывать УИК)

```python
disp = dict()
for region in df['Регион'].unique():
    disp[region] = ((df[df['Регион'] == region][m2_col] - df[df['Регион'] == region][m2_col].mean())**2).sum() / (len(df[df['Регион'] == region]) - 1)

disp_df = pd.Series(disp)
disp_df[:10]
```

```python
res = np.zeros((len(candidates), 101))
for i1, name in enumerate(candidates):
    perc = np.round(100 * df[name] / df[m2_col])
    perc_count = perc.value_counts()
    for i2 in perc_count.index:
        res[i1, int(i2)] = perc_count[i2]
```

```python
pd.set_option('display.max_columns', None)
pd.DataFrame(res, columns=range(0, 101), index=candidates)
```
